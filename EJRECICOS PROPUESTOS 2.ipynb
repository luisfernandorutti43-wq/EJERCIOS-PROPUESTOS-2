```python
import pandas as pd

inicio = int(input("Ingrese el año de inicio: "))
fin = int(input("Ingrese el año de fin: "))

ventas = {}
for año in range(inicio, fin + 1):
    venta = float(input(f"Ventas del año {año}: "))
    ventas[año] = venta

serie_ventas = pd.Series(ventas)
print("\nVentas originales:")
print(serie_ventas)

serie_descuento = serie_ventas * 0.9
print("\nVentas con 10% de descuento:")
print(serie_descuento)
```

```python
import pandas as pd

def resumen_notas(dic_notas):
    serie = pd.Series(dic_notas)
    resumen = pd.Series({
        'Mínima': serie.min(),
        'Máxima': serie.max(),
        'Media': serie.mean(),
        'Desviación Típica': serie.std()
    })
    return resumen

# Ejemplo:
notas = {'Ana': 15, 'Luis': 12, 'Pedro': 18, 'Marta': 10}
print(resumen_notas(notas))
```

```python
import pandas as pd

def notas_aprobadas(dic_notas):
    serie = pd.Series(dic_notas)
    aprobados = serie[serie >= 11].sort_values(ascending=False)
    return aprobados

# Ejemplo:
notas = {'Ana': 15, 'Luis': 12, 'Pedro': 8, 'Marta': 10}
print(notas_aprobadas(notas))
```

```python
def balance_total(df, meses):
    df_filtrado = df[df['Mes'].isin(meses)]
    df_filtrado['Balance'] = df_filtrado['Ventas'] - df_filtrado['Gastos']
    total_balance = df_filtrado['Balance'].sum()
    return total_balance

# Ejemplo de uso:
meses_consultar = ['Enero', 'Marzo']
total = balance_total(df, meses_consultar)
print(f"Balance total en {meses_consultar}: {total}")
```

```python
import pandas as pd

def resumen_cotizaciones():
    # Simulando cotizacion.csv con un diccionario
    datos = {
        'Nombre': ['EmpresaA', 'EmpresaB', 'EmpresaC'],
        'Final': [10.5, 15.2, 8.7],
        'Máximo': [11.0, 16.0, 9.0],
        'Mínimo': [10.0, 14.5, 8.5],
        'Volumen': [1500000, 2300000, 1250000],
        'Efectivo': [15750, 34960, 10875]
    }

    # Crear DataFrame
    df = pd.DataFrame(datos)
    print("DataFrame original:\n", df)

    # Cálculos de resumen (mínimo, máximo, media)
    resumen = pd.DataFrame({
        'Mínimo': df[['Final', 'Máximo', 'Mínimo', 'Volumen', 'Efectivo']].min(),
        'Máximo': df[['Final', 'Máximo', 'Mínimo', 'Volumen', 'Efectivo']].max(),
        'Media': df[['Final', 'Máximo', 'Mínimo', 'Volumen', 'Efectivo']].mean()
    })

    return resumen

# Mostrar resumen
print("\nResumen de cotizaciones:")
print(resumen_cotizaciones())
```

```python
import pandas as pd
import numpy as np
from datetime import datetime

# ----------------------------
# Simular datos como si fueran de los archivos CSV
# ----------------------------
data = {
    'ESTACION': [1, 1, 2, 2, 3, 3],
    'MAGNITUD': [10, 20, 10, 20, 10, 20],
    'ANO': [2016, 2016, 2017, 2017, 2018, 2019],
    'MES': [1, 1, 2, 2, 3, 4],
    'D01': [50, 80, 55, 70, 60, 75],
    'D02': [45, 85, 50, 75, 65, 80]
}

df = pd.DataFrame(data)

# ----------------------------
# Filtrar columnas necesarias
# ----------------------------
columnas_dias = [col for col in df.columns if col.startswith('D')]
df_filtrado = df[['ESTACION', 'MAGNITUD', 'ANO', 'MES'] + columnas_dias]

# ----------------------------
# Reestructurar (melt) para pasar columnas D01, D02... a filas
# ----------------------------
df_melted = df_filtrado.melt(id_vars=['ESTACION', 'MAGNITUD', 'ANO', 'MES'],
                             var_name='DIA', value_name='VALOR')

# ----------------------------
# Crear columna FECHA
# ----------------------------
df_melted['DIA'] = df_melted['DIA'].str.extract('(\d+)').astype(int)
df_melted['FECHA'] = pd.to_datetime(dict(year=df_melted['ANO'],
                                         month=df_melted['MES'],
                                         day=df_melted['DIA']), errors='coerce')

# ----------------------------
# Eliminar fechas no válidas (NaT)
# ----------------------------
df_melted = df_melted[~df_melted['FECHA'].isna()]

# ----------------------------
# Ordenar DataFrame
# ----------------------------
df_melted = df_melted.sort_values(by=['ESTACION', 'MAGNITUD', 'FECHA'])

# ----------------------------
# Mostrar Estaciones y Contaminantes
# ----------------------------
print("\nEstaciones disponibles:", df_melted['ESTACION'].unique())
print("Contaminantes disponibles:", df_melted['MAGNITUD'].unique())

# ----------------------------
# Función: Filtrar emisiones por estación, contaminante y rango de fechas
# ----------------------------
def emisiones_filtradas(estacion, contaminante, fecha_inicio, fecha_fin):
    mask = (
        (df_melted['ESTACION'] == estacion) &
        (df_melted['MAGNITUD'] == contaminante) &
        (df_melted['FECHA'] >= pd.to_datetime(fecha_inicio)) &
        (df_melted['FECHA'] <= pd.to_datetime(fecha_fin))
    )
    return df_melted.loc[mask, ['FECHA', 'VALOR']]

# ----------------------------
# Resumen Descriptivo General por Contaminante
# ----------------------------
print("\nResumen descriptivo por Contaminante:")
print(df_melted.groupby('MAGNITUD')['VALOR'].describe())

# ----------------------------
# Resumen Descriptivo por Contaminante y Estación (Distritos)
# ----------------------------
print("\nResumen descriptivo por Contaminante y Estación:")
print(df_melted.groupby(['ESTACION', 'MAGNITUD'])['VALOR'].describe())

# ----------------------------
# Función: Resumen descriptivo de emisiones por estación y contaminante
# ----------------------------
def resumen_estacion_contaminante(estacion, contaminante):
    mask = (df_melted['ESTACION'] == estacion) & (df_melted['MAGNITUD'] == contaminante)
    return df_melted.loc[mask, 'VALOR'].describe()

# ----------------------------
# Función: Emisiones medias mensuales de un contaminante en un año
# ----------------------------
def emisiones_media_mensual(contaminante, año):
    mask = (df_melted['MAGNITUD'] == contaminante) & (df_melted['ANO'] == año)
    return df_melted.loc[mask].groupby('MES')['VALOR'].mean()

# ----------------------------
# Función: Medias mensuales de todos los contaminantes de una estación
# ----------------------------
def medias_mensuales_estacion(estacion):
    mask = (df_melted['ESTACION'] == estacion)
    return df_melted.loc[mask].groupby(['MAGNITUD', 'MES'])['VALOR'].mean().unstack()

# ----------------------------
# Ejemplos de uso:
# ----------------------------

print("\nEmisiones filtradas estación 1, contaminante 10, fechas 2016-01-01 a 2016-01-31:")
print(emisiones_filtradas(1, 10, '2016-01-01', '2016-01-31'))

print("\nResumen de emisiones estación 2, contaminante 20:")
print(resumen_estacion_contaminante(2, 20))

print("\nEmisiones medias mensuales del contaminante 10 en el año 2016:")
print(emisiones_media_mensual(10, 2016))

print("\nMedias mensuales de contaminantes en estación 1:")
print(medias_mensuales_estacion(1))

```
